```python
training_annotations_filepath = "/tmp/akshett.jindal/annotations/instances_train2017.json"
validation_annotations_filepath = "/tmp/akshett.jindal/annotations/instances_val2017.json"
```

```python
training_annotations = COCO(training_annotations_filepath)
validation_annotations = COCO(validation_annotations_filepath)
```

```python
coco_categories = training_annotations.cats
list(coco_categories.items())[0]
```

```python
all_supercategories = set(cat['supercategory'] for cat in coco_categories.values())
all_categories = set(cat['name'] for cat in coco_categories.values())
```

```python
all_supercategories
```

```python
all_categories
```

```python
image_objects = []

for image_id, row in stiminfo.iterrows():
    coco_image_id, coco_split = row['cocoId'], row['cocoSplit']
    annotations = training_annotations if coco_split == "train2017" else validation_annotations
    categories = set([coco_categories[value['category_id']]['name'] for value in annotations.imgToAnns[coco_image_id]])
    super_categories = set([coco_categories[value['category_id']]['supercategory'] for value in annotations.imgToAnns[coco_image_id]])

    image_objects.append({
        "nsd_image_id": image_id,
        "coco_image_id": coco_image_id,
        "coco_split": coco_split,
        "categories": list(categories),
        "supercategories": list(super_categories),
    })
```

```python
objects_df = pd.DataFrame(image_objects)
objects_df
```

```python
objects_df.to_pickle('/tmp/akshett.jindal/nsddata_objects.pkl')
```
