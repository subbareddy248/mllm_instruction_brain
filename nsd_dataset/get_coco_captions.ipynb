```python
import pandas as pd
from pycocotools.coco import COCO
```

```python
!wget "http://images.cocodataset.org/annotations/annotations_trainval2017.zip" -O "/tmp/akshett.jindal/annotations_trainval2017.zip"
```

```python
training_annotations_filepath = "/tmp/akshett.jindal/annotations/captions_train2017.json"
validation_annotations_filepath = "/tmp/akshett.jindal/annotations/captions_val2017.json"
```

```python
training_annotations = COCO(training_annotations_filepath)
validation_annotations = COCO(validation_annotations_filepath)
```

```python
print(training_annotations.__dict__.keys())
print(dir(training_annotations))
```

```python
list(training_annotations.imgToAnns.items())[:5]
```

```python
set(len(value) for key, value in training_annotations.imgToAnns.items())
```

```python
image_captions = []

for image_id, row in stiminfo.iterrows():
    coco_image_id, coco_split = row['cocoId'], row['cocoSplit']
    # print(f"{image_id = }, {coco_image_id = }, {coco_split = }")
    captions = None
    if coco_split == "train2017":
        captions = [value['caption'] for value in training_annotations.imgToAnns[coco_image_id]]
    else:
        # Must be from validation
        captions = [value['caption'] for value in validation_annotations.imgToAnns[coco_image_id]]
    # print(captions)

    image_captions.append({
        'nsd_image_id': image_id,
        'coco_image_id': coco_image_id,
        'coco_split': coco_split,
        'captions': captions,
    })
```

```python
captions_df = pd.DataFrame(image_captions)
captions_df
```

```python
captions_df.to_pickle('/tmp/akshett.jindal/nsddata_captions.pkl')
```
